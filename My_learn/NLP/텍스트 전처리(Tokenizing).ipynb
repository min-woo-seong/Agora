<자연어 처리 개요>
-텍스트 분석의 종류
1. 자연어 : 한국어,영어
2. 인공어 : 자연어를 가공하여 만든언어
3. 통제자연어 : 기계가 이해하기 쉬운 자연어
4. 형식어 : 프로그래밍 언어

-텍스트 분석의 절차
1. corpus(말뭉치) 텍스트 수집
2. 전처리 : 결측값이나 불필요한 표현 삭제 
3. 정규화 : 분석에 편리한 형태로 변환 
    3.1 : 토큰화 : 형태소나 단어롤 분리
    3.2. lemmatization : 사전에 등재된 형태로 바꿈
           stemming : 휴리스틱을 이용 어간을 분리

-텍스트 분석의 종류
: 요약, 감성분석, 관계추출, 상식추론, 질의답변, 기계변역 등

-텍스트 분석의 문제점
: 자연어처리의 발전으로 점점 복잡해짐
  그로인한 데이터의 부족현상
  데이터가 편향 되어 있다면 학습한 패턴도 편향 되어짐

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

text = "I am actively looking for Ph.D. students. and you are a Ph.D. student."
tokenized_sentence = word_tokenize(text)
```

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma
```

```python
okt = Okt()
kkma = Kkma()
```

```python
okt.morphs("열심히 코딩한 당신, 연휴에는 여행을 가라")
```

```python
okt.pos("열심히 코딩한 당신, 연휴에는 여행을 가라")
```

```python
okt.nouns("열심히 코딩한 당신, 연휴에는 여행을 가라")
```

```python
print('꼬꼬마 형태소 분석 :', kkma.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('꼬꼬마 품사 태깅 :', kkma.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('꼬꼬마 명사 추출 :', kkma.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```

##어간 추출(Stemming)과 표제어 추출(Lemmatization)

```python
from nltk.stem import WordNetLemmatizer
n= WordNetLemmatizer()
words = ['policy', 'doing','organization','have','going','love','lives','fly','dies','watched','has','starting']
print([n.lemmatize(w) for w in words])
```

```python
n.lemmatize('dies', 'v')
```

```python
n.lemmatize('watched','v')
```

```python
n.lemmatize('has','v')
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()
```

```python
sentence = "This was not the map we found in Billy Bones's chest, but an accurate copy, complete in all things--names and heights and soundings--with the single exception of the red crosses and the written notes."
```

```python
tokenized_sentence = word_tokenize(sentence)
```

```python
print('어간 추출전 : ', tokenized_sentence)
print('어간 추출후 : ', [stemmer.stem(word) for word in tokenized_sentence])
```

```python
words = ['formalize', 'allowance', 'electricical']
words
```

```python
[stemmer.stem(word)for word in words]
```
