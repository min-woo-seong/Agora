```python
import matplotlib.pyplot as plt
```

```python
# x축의 입력값 (1,2,3), y축의 입력값 (2,4,6)으로 선 그래프 작성
plt.plot([1,2,3],[2,4,6])

# 선 그래프의 타이틀 Hello plot설정
plt.title("Hello plot")

# 선 그래프 출력
plt.show()
```

## pyplot의 2가지 요소 -  Figure , Axes
### Figure : 그림을 그릴때 캔버스 역할을 담당, 그림판의 크기를 조절 한다거나 플롯을 최종화면에 나타내는 역할
### Axes : 실제 그림을 그리는 매서드를 가고 X축, Y축, 타이틀, 범례등의 속성설정을 할수있음


```python
# plt.figure()는 주로 figure의 크기를 조절하는 데 사용됨.
plt.figure(figsize=(10,4))

plt.plot([1,2,3],[2,4,6])
plt.title("Hello plot")
plt.show()
```

```python
figure = plt.figure(figsize=(10,4))
print(type(figure))
```

```python
# figure 그림판의 색깔을 바꾸기 facecolor
plt.figure(figsize=(4,4), facecolor='gray')
plt.plot([1,2,3],[2,4,6])
plt.title("Hello plot")
plt.show()
```

```python
ax = plt.axes()
print(type(ax))
```

```python
# pyplot 에서 설정된 figure와 axes 객체 함께 가져오기
fig, ax = plt.subplots()
print('fig type:', type(fig), '\nax type:', type(ax))
```

```python
fig, ax = plt.subplots(figsize=(4,4))

# axes객체의 plot()을 이용 선그래프를 그림
ax.plot([1,2,3],[2,4,6])

# axes 객체의 set_title()을 이용 선 그래프 타이틀을 hello plot로 설정
ax.set_title('Hello plot')
plt.show()
```

```python
# 여러개의 plot을 가지는 subplot들을 생성하기
# 두개의 axes를 생성 전체 크기가 가로6, 세로 3인 figure생성
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
```
